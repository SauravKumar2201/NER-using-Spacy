##Named Entity Recognition Model
Named entity Recognition using spacy model

Import required library

```
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
```

Test sentence

```
sentence = "satyam is the CEO of Apple and he live in USA"
```

Creating a list of tuples generated from model

```
doc = nlp(sentence)
hello =([(X.text, X.label_) for X in doc.ents])
```

Function to parse list of tuples generated by model

```
def entity_name(name):
    for entity in hello:
        if entity[1] == name:
            print(entity[0])
```

Results

```
print("\nOrganization(s): ")
entity_name("ORG")
print("\nPerson(s):")
entity_name("PERSON")
print("\nLocation(s):")
entity_name("GPE")
```
